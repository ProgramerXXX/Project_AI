```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.random.rand(40)
y = 0.5 * x + 0.2 + 0.1*np.random.rand(40)
plt.plot(x,y,'.')
plt.show()
```

```python
import tensorflow as tf
```

```python
w = tf.Variable(np.random.rand())
b = tf.Variable(np.random.rand())
lr = 0.1
for epoch in range(1000):
    with tf.GradientTape() as t:
        Y = w * x + b
        loss = tf.reduce_mean((Y - y)**2)
    dw , db = t.gradient(loss,[w,b])
    w.assign_sub(lr*dw)
    b.assign_sub(lr*db)
    print(epoch,w.numpy(),b.numpy(),loss.numpy())
```

```python
Z = w * x + b
plt.plot(x,y,'.b')
plt.plot(x,Z,'.r')
plt.show()
```

Use Libary Keras in Tensorflow 


```python
d_in = (1,)
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1,input_shape=d_in))
model.compile(loss=tf.keras.losses.MeanSquaredError(),
            optimizer=tf.keras.optimizers.SGD(learning_rate=0.1))
model.fit(x,y,epochs=999)

z = model.predict(x)
plt.plot(x,y,'.b')
plt.plot(x,z,'.r')
plt.show()
```
