```python
import pandas as pd 
import numpy as np 
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_csv('let.csv')
data.head()
```

```python
X = data.iloc[:,1:]
X.head()
```

```python
Y = data.iloc[:,0]
Y.head()
```

```python
modelSVM = SVC(kernel='poly')
modelTree = DecisionTreeClassifier()
Svm = cross_val_score(modelSVM,X,Y,cv=5)
Tree = cross_val_score(modelTree,X,Y,cv=5)
print(Svm.round(4))
print(Tree.round(4))
```
